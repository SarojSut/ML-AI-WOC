```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv("D:\DataSets\multi_classification_train.csv")
df_test = pd.read_csv("D:\DataSets\multi_classification_test.csv")
df_train = df_train.drop(columns=["ID"])
df_test = df_test.drop(columns=["ID"])
df_train.head(50)
```

```python
X_train = df_train.iloc[:,0:20].values
Y_train = df_train.iloc[:,20].values

test = df_test.iloc[:,:].values

print(f"Trainng set shape: {X_train.shape}")
print(f"Shape of class(labels) in Training set: {Y_train.shape}")
print(f"Maximum of the Training set: {np.max(X_train)}")
print(f"Minimum of the Training set: {np.min(X_train)}")
print(f"Test set shape: {X_train.shape}")
```

Normilaziation of Data makes all the data to fall in certain interval.
It improve perfomance, avoid dominance of datas which lies under certain range, also maintains numerical stability

```python
def normalize(x):
    x_mean = x.mean()
    x_std = x.std()
    x_normal = (x-x_mean)/x_std
    return x_normal

X_train_normal = normalize(X_train)
```

sigmoid(): is an activation function for output of hidden layers

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

softmax(): activation function for final output i.e. last layer
Make sure to maintain the proper Dimension in softmax activation

```python
def softmax(z):
    z_max = np.max(z, axis=1, keepdims=True)
    exp_z = np.exp(z - z_max)
    return exp_z/(np.sum(exp_z, axis=1, keepdims=True))
```

sigmoid_drivative(): return the derivative of sigmoid function.
Usefull in back propogation

```python
def sigmoid_derivative(z):
    return sigmoid(z)*(1 - sigmoid(z))
```

cross_entropy_loss(): it is the softmax's loss function which calculates and return loss

```python
def cross_entropy_loss(y_true, y_pred):
    y_pred = np.where(y_pred<=1e-12, 1e-12, y_pred)
    loss = np.sum(y_true*np.log(y_pred), axis=1)
    cost = -np.mean(loss)
    return cost
```

One Hot Encoding 
Basic purpose of this is to convert categorical data into numerical data.
Here in this case I have 3 different categories of FlowerType, so Hot Encoding to creat a matrix of shape=(len(X_train),Unique Categories).
Also initalizing the matrix to represent different categories into Binary Vector format.

```python
def one_hot_encode(labels):
    labels = labels.astype(int)
    n_labels = len(labels)
    n_unique_labels= len(np.unique(labels))
    one_hot_encode = np.zeros((n_labels, n_unique_labels))
    for i in range(n_labels):
        one_hot_encode[i,labels[i]] = 1
    return one_hot_encode

Y_train_one_hot = one_hot_encode(Y_train)
```

Below is the basic Neural Network Architecture

input layer has 20 units
1st layer has 32 Neurons (its 1st hidden layer)
2nd layer has 16 neurons (its 2nd hidden layer)
3rd layer has 5 neurons (its the output layer)

Weights Are initalized randomly
Each weight is multipied by squareRoot of (1/no. of preciding neurons) it is just a better way of initalization.
w1 : weights used in 1st layer shape(20,32)
w2 : weights used in 2nd layer shape(32,16)
w3 : weights used in 3rd layer shape(16,5)

```python
m = len(X_train)
n_inputs  = len(X_train[0])
a1 = 32
a2 = 16
a3 = 5

w1 = np.random.rand(n_inputs, a1)*np.sqrt(1/n_inputs)
w2 = np.random.rand(a1, a2)*np.sqrt(1/a1)
w3 = np.random.rand(a2, a3)*np.sqrt(1/a2)

b1 = np.zeros((m, a1))
b2 = np.zeros((m, a2))
b3 = np.zeros((m, a3))
```

Next cell is implementation of model.
Converting Architecture to mathematics and forming the Neural Netwok

```python
epochs = 999
learning_rate = 0.01
batch_size = 1000

for i in range(epochs):
    for j in range(0, len(X_train), batch_size):
        batch_x = X_train[j:j+batch_size]
        batch_y = Y_train_one_hot[j:j+batch_size]
        batch_Y_train = Y_train[j:j+batch_size]
        
        #forward propogation
        z1 = np.dot(batch_x, w1) + b1[j:j+batch_size]
        a1 = sigmoid(z1)
        z2 = np.dot(a1, w2) + b2[j:j+batch_size]
        a2 = sigmoid(z2)
        z3 = np.dot(a2, w3) + b3[j:j+batch_size]
        a3 = softmax(z3)

        #backword propogation
        err_a3 = a3 - batch_y
        err_a2 = err_a3.dot(w3.T)*sigmoid_derivative(a2)
        err_a1 = err_a2.dot(w2.T)*sigmoid_derivative(a1)

        #weights updation
        w3 -= learning_rate*(a2.T).dot(err_a3)
        w2 -= learning_rate*(a1.T).dot(err_a2)
        w1 -= learning_rate*(batch_x.T).dot(err_a1)

        #biaes updation
        b3[j:j+batch_size] -= learning_rate*err_a3
        b2[j:j+batch_size] -= learning_rate*err_a2
        b1[j:j+batch_size] -= learning_rate*err_a1

    loss = cross_entropy_loss(batch_y, a3)
    prediction = np.argmax(a3, axis=1)
    accuracy = np.mean(prediction==batch_Y_train)*100
    print(f"Epoch: {1+i} Loss: {loss} Accuracy: {accuracy}")

```

```python
#printing weights and biases
print(f"Weight 1: {w1}\nWeight 2: {w2}\nWeight 3: {w3}")
print(f"Bias 1: {b1}\nBias 2: {b2}\n Bias 3: {b3}")
```

Next cell is the Evaluation.
Trying the trained weights on the test set and recording the accuracy!!!

```python
a1 = sigmoid(np.dot(test,w1) + b1[:len(test)])
a2 = sigmoid(np.dot(a1, w2) + b2[:len(test)])
a3 = softmax(np.dot(a2, w3) + b3[:len(test)])
test_prediction = np.argmax(a3, axis=1)

print(f"Shape of prediction: {test_prediction.shape}")
print(f"Test Prediction: {test_prediction}")
print(f"1st 50 prediction on test set: {test_prediction[:50]}")
```
